```python
import cv2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
image = cv2.imread("dollar2.png",0)
plt.imshow(image,'gray')
```

```python
b0 = (image >> 0) & 1
b1 = (image >> 1) & 1
b2 = (image >> 2) & 1
b3 = (image >> 3) & 1
b4 = (image >> 4) & 1
b5 = (image >> 5) & 1
b6 = (image >> 6) & 1
b7 = (image >> 7) & 1
```

```python
plt.figure(figsize=(60,10))
fig=plt.figure(dpi=200)

plt.subplot(3,3,1)
plt.imshow(b0,cmap='gray')
plt.axis("off")
plt.title("bit b0")

plt.subplot(3,3,2)
plt.imshow(b1,cmap='gray')
plt.axis("off")
plt.title("bit b1")

plt.subplot(3,3,3)
plt.imshow(b2,cmap='gray')
plt.axis("off")
plt.title("bit b2")

plt.subplot(3,3,4)
plt.imshow(b3,cmap='gray')
plt.axis("off")
plt.title("bit b3")

plt.subplot(3,3,5)
plt.imshow(b4,cmap='gray')
plt.axis("off")
plt.title("bit b4")

plt.subplot(3,3,6)
plt.imshow(b5,cmap='gray')
plt.axis("off")
plt.title("bit b5")

plt.subplot(3,3,7)
plt.imshow(b6,cmap='gray')
plt.axis("off")
plt.title("bit b6")

plt.subplot(3,3,8)
plt.imshow(b7,cmap='gray')
plt.axis("off")
plt.title("bit b7")

plt.subplot(3,3,9)
plt.imshow(image,cmap='gray')
plt.axis("off")
plt.title(" orignal img")

```

```python
c= b6+b7
plt.imshow(c,cmap='gray')
plt.axis("off")
plt.title("image after adding bit plane b6 and b7")
```

```python
c= b6+b7+image
plt.imshow(c,cmap='gray')
plt.axis("off")
plt.title("image after adding bit plane b6 and b7 and orignal image")
```

```python
c= b5+b6+b7
plt.imshow(c,cmap='gray')
plt.axis("off")
plt.title("image after adding bit plane b5, b6 and b7")
```

```python
c= b3+b4+b5
plt.imshow(c,cmap='gray')
plt.axis("off")
plt.title("image after adding bit plane b4, b5 and b6")
```

```python
c= b0+b1+b2
plt.imshow(c,cmap='gray')
plt.axis("off")
plt.title("image after adding bit plane b0, b1 and b2")
```
