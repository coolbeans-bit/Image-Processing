```python
import cv2
import numpy as np
import matplotlib. pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')
```

```python
image=cv2.imread("cat.jpeg")
```

```python
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
print(image.dtype) 
print(image.shape)
```

```python
if image is None:
 raise ValueError("Image not loaded correctly. Check the file path.")

# Convert the image to grayscale if it's not already
if len(image.shape) == 3:
 image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
equalized_image = cv2.equalizeHist(image)
```

```python
plt.figure(figsize=(10, 5))
```

```python
cv2.imwrite('equalized_image.jpg', equalized_image)
```

```python
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title('original image')
plt.subplot(1,2,2)
plt.imshow (equalized_image, cmap='gray')
plt.title('Histogram equilized image')
```

```python
plt.figure(figsize=(10, 5))
plt.hist(image.ravel(), bins=256, range=[0, 256], color="gray")
plt.title('Histogram of Original Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(equalized_image.ravel(), bins=256, range=[0, 256], color="gray")
plt.title('Histogram of Equalized Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
bright_image = cv2.add (image, 150)
plt.imshow(bright_image, cmap = 'gray')
plt.title('Brighten image' )
#plt.axis('off')
plt.show()
```

```python
equalized_image1 = cv2.equalizeHist(bright_image)
```

```python
plt.subplot(1,2,1)
plt.imshow(bright_image, cmap='gray')
plt.title('brighten image')
plt.subplot(1,2,2)
plt.imshow (equalized_image1, cmap='gray')
plt.title('Equilized brighten image')
```

```python
plt.figure(figsize=(10, 5))
plt.hist(bright_image.ravel(), bins=256, range=[0, 256], color="gray")
plt.title('Histogram of brightrn image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(equalized_image1.ravel(), bins=256, range=[0, 256], color="gray")
plt.title('Histogram of equlized bright image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```
