```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image=cv2.imread("grayscale.jpg")
# Display the original image
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
negative_image = 255 - image

plt.imshow(negative_image, cmap='gray')
plt.title('Negative Image')
plt.axis('off')
plt.show()
```

```python
bright_image = cv2.add(image, 100)    
plt.imshow(bright_image, cmap='gray')
plt.title('Brightness Improved')
plt.axis('off')
plt.show()
```

```python
def decrease_brightness(image, value=30):
    # Subtract the value from all pixels in the image
    dark_image = cv2.subtract(image, value)
    return dark_image

# Apply brightness reduction
dark_image = decrease_brightness(image, 200)

# Display the brightness reduced image
plt.imshow(dark_image, cmap='gray')
plt.title('Brightness Reduced')
plt.axis('off')
plt.show()
```

```python
# Apply binary thresholding
threshold = 128
_, thresh_image = cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY)

# Display the thresholded image
plt.imshow(thresh_image, cmap='gray')
plt.title('Thresholded Image')
plt.axis('off')
plt.show()
```
